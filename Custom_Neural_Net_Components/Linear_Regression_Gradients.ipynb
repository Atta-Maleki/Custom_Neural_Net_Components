```
hello = tf.ones((2,2))

with tf.GradientTape(persistent = True) as gt:
  gt.watch(hello)
  shello= tf.reduce_sum(hello)
  fhello = tf.square(shello)
dfhello_dhello  = gt.gradient(fhello, hello)
print(dfhello_dhello)
```

```
var_x = tf.constant(3.0)
with tf.GradientTape(persistent = True) as gt:
  gt.watch(var_x)
  var_y = var_x * var_x
  var_z = var_y * var_y
dz_dx = gt.gradient(var_z, var_x)
print(dz_dx)
dy_dx = gt.gradient(var_y, var_x)
print(dy_dx)
```

## Nested Gradient

```
vx = tf.Variable(3)
with tf.GradientTape() as gt:
  with tf.GradientTape() as gt2:
    vy = vx * vx * vx
  dr_df = gt2.gradient(vy, vx)
d2y_dx2 = gt.gradient(dr_df, vx)

print(dy_dx)
print(d2y_dx2)
```
