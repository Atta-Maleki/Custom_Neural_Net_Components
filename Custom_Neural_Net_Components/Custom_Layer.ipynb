```
import tensorflow as tf
import numpy as np
```

```
xs = np.array([5.0, 6.0, 7.0, 8.0 ,9.0, 10.0], dtype=float)
ys = np.array([9.0, 11.0, 13.0, 15.0, 17.0, 19.0], dtype=float)
```

```
print(xs.shape)
```

```
xs = np.reshape(xs, (-1,1))
print(xs.shape)
```

```
class myLayerClass(tf.keras.layers.Layer):
  def __init__(self, units = 32):
    super(myLayerClass, self).__init__()
    self.units = units
  
  def build(self, input_shape):
    w_init = tf.random_normal_initializer()
    self.w = tf.Variable(name="nodeWeights",
                         initial_value=w_init(shape=(input_shape[-1], self.units),
                                              dtype = 'float32'),
                          trainable = True)
    b_init = tf.zeros_initializer()
    self.b = tf.Variable(name = "nodeBias", 
                         initial_value=b_init(shape=(self.units,), 
                                                dtype = 'float32'), 
                         trainable = True)
  def call(self, inputs):
    return tf.matmul(inputs, self.w)+self.b
```

```
example = myLayerClass(units = 1)
x = tf.ones((1,1))
y = example(x)
print(example.variables)
```

```
myLayer = myLayerClass(units=1)
```

```
model = tf.keras.Sequential([myLayer])
model.compile(optimizer='sgd', loss = 'mean_squared_error')
model.fit(xs, ys, epochs = 550, verbose = 0)
print(model.predict([[4.0]]))
```
